```python
def convert_labels(path, x1, y1, x2, y2):
    """
    Definition: Parses label files to extract label and bounding box
        coordinates.  Converts (x1, y1, x1, y2) format to
        (x, y, width, height) normalized YOLO format.
    """
    def sorting(l1, l2):
        if l1 > l2:
            lmax, lmin = l1, l2
            return lmax, lmin
        else:
            lmax, lmin = l2, l1
            return lmax, lmin
    dir_name,img_name=os.path.split(path)
    print(img_name)
    print(dir_name)
    img_label_name=img_name[:len(img_name) - 4]+'.txt'
    img_label_name = dir_name + "/" + img_label_name
    print(img_label_name)
    size = get_img_shape(path)
    xmax, xmin = sorting(x1, x2)
    ymax, ymin = sorting(y1, y2)
    dw = 1./size[1]
    dh = 1./size[0]
    x = (xmin + xmax)/2.0
    y = (ymin + ymax)/2.0
    w = xmax - xmin
    h = ymax - ymin
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    with open(img_label_name,"w")as f:
        label_str="0"+" "+str(x)+" "+str(y)+" "+str(w)+" "+str(h)+"\n"
        print(label_str)
        f.write(label_str)
```

```python
def get_img_shape(path):
    img = cv2.imread(path)
    try:
        print(img.shape)
        return img.shape
    except AttributeError:
        print('error! ', path)
        return (None, None, None)
```

```python
for j in range(df2.shape[0]):
    img_path=img_path_dir_name+df1[j]
    x1=df2[j:j+1]['x1'].values[0]
    y1=df2[j:j+1]['y1'].values[0]
    x2=df2[j:j+1]['x2'].values[0]
    y2=df2[j:j+1]['y2'].values[0]
    convert_labels(img_path,x1,y1,x2,y2)
```
